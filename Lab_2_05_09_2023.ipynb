```python
# 4. For each of the orgs in all-orgs-strings.lc.txt, determine how often they occur in the corpus
print("4. For each of the orgs in all-orgs-strings.lc.txt, determine how often they occur in the corpus")
nltk.download("punkt")

with open("./all-orgs-strings.lc.txt", "r") as org_file:
    orgs = [line.strip() for line in org_file]

# Tokenize org names using NLTK
orgs_tokens = [nltk.word_tokenize(org) for org in orgs]
```

```python
# 4. For each of the orgs in all-orgs-strings.lc.txt, determine how often they occur in the corpus
import os

corpus_directory = "./reuters21578"
corpus_files = [os.path.join(corpus_directory, filename) for filename in os.listdir(corpus_directory)]

org_counts = {}

for org in orgs_tokens:
    org_count = 0
    for corpus_file in corpus_files:
        # corpus file ending with .sgm only
        if (corpus_file.endswith(".sgm")):
            with open(corpus_file, "r") as file:
                raw = file.read()
                # Tokenize the text in the corpus
                corpus_tokens = nltk.word_tokenize(raw)
                # Use NLTK's FreqDist to count occurrences of org in the corpus
                org_count += nltk.FreqDist(corpus_tokens)[org[0]]  # Assuming org is a single word
            org_counts[' '.join(org)] = org_count
```

```python
for org, count in org_counts.items():
    print(f"{org}: {count} occurrences")
```
